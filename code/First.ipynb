# FS데이터와 PMOD데이터 합치기 -> total 데이터 생성

```python
fs_lt = FS.columns
pmod_lt = PMOD.columns
```

```python
fs_lt = set(fs_lt)
pmod_lt = set(pmod_lt)

intersec=list(pmod_lt&fs_lt)
intersec
```

```python
base = FS.iloc[:,:15]
base
```

```python
FS_pet = FS.loc[:,['FC', 'LTC', 'PC', 'OC', 'GCA', 'GCP', 'PQ','MTC', 'CN', 'PUT', 'THA', 'Composite*']]
FS_pet
```

```python
PMOD_pet = PMOD.loc[:,['FC', 'LTC', 'PC', 'OC', 'GCA', 'GCP', 'PQ','MTC', 'CN', 'PUT', 'THA', 'Composite*']]
PMOD_pet
```

```python
PET = pd.DataFrame()
PET['ID'] = FS['ID']
intersec = ['FC', 'LTC', 'PC', 'OC', 'GCA', 'GCP', 'PQ','MTC', 'CN', 'PUT', 'THA', 'Composite*']

for col in intersec:
    PET[col] = (FS_pet[col] + PMOD_pet[col])/2

PET['ID'] = PET['ID'].astype('int')
PET
```

```python
total = pd.merge(base, PET, on='ID')
total
```

```python
total = pd.merge(total, FS[['ID','BS', 'CBL']], on='ID')
total
```

```python
total = pd.merge(total, PMOD[['ID','CBLCTX', 'Mid', 'Pons']], on='ID')
total
```

```python
total.info()
```

```python
total.to_csv('../알츠하이머/data/total.csv', index=False)
```

# Onset age 결측치 처리

```python
import pandas as pd
```

```python
data.columns
```

## 결측치 0으로 처리

```python
data = pd.read_csv('../알츠하이머/data/total.csv')
data['Onset age'] = data['Onset age'].fillna(0)
data[['positivity','APOE','Age','Onset age','Diagnosis code']].corr()
```

## 결측치 평균으로 처리

```python
data = pd.read_csv('../알츠하이머/data/total.csv')
data['Onset age'] = data['Onset age'].fillna(data['Onset age'].mean())
data[['positivity','APOE','Age','Onset age','Diagnosis code']].corr()
```

## 결측치 중앙값으로 처리

```python
data = pd.read_csv('../알츠하이머/data/total.csv')
data['Onset age'] = data['Onset age'].fillna(data['Onset age'].median())
data[['positivity','APOE','Age','Onset age','Diagnosis code']].corr()
```

## 결측치 회귀로 처리

```python
data = pd.read_csv('../알츠하이머/data/total.csv')
input_data = data[data['Onset age'].isnull()==False]
target_data = data[data['Onset age'].isnull()==True]
```

```python
data
```

```python
train_x = input_data.drop(['ID','Onset age'], axis=1)
train_y = input_data['Onset age']
test_x = target_data.drop(['ID','Onset age'], axis=1)
```

```python
print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
```

```python
input_data.corr().iloc[4].sort_values(ascending=False)
```

### 선형회귀

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(train_x[['Age', 'BS', 'SNSB']], train_y)
```

```python
pred_lr = lr.predict(test_x[['Age', 'BS', 'SNSB']])
pred_lr = pred_lr.astype('int')
pred_lr = pred_lr.tolist()
lt = list(target_data.index)
```

```python
data = pd.read_csv('../알츠하이머/data/total.csv')
n=0
lt = list(target_data.index)

for _ in range(len(pred_lr)):
    data.loc[lt[n],'Onset age'] = pred_lr[n]
    n += 1 
```

```python
data.isnull().sum()
```

```python
data[['positivity','APOE','Age','Onset age','Diagnosis code']].corr()
```

### XGBoost

```python
from xgboost.sklearn import XGBRegressor
xgb = XGBRegressor()
xgb.fit(train_x[['Age', 'BS', 'SNSB']], train_y)
```

```python
pred_xgb = xgb.predict(test_x[['Age', 'BS', 'SNSB']])
pred_xgb = pred_xgb.astype('int')
pred_xgb = pred_xgb.tolist()
lt = list(target_data.index)
```

```python
data = pd.read_csv('../알츠하이머/data/total.csv')
n=0
lt = list(target_data.index)

for _ in range(len(pred_xgb)):
    data.loc[lt[n],'Onset age'] = pred_xgb[n]
    n += 1 
```

```python
data.isnull().sum()
```

```python
data[['Onset age','positivity','Diagnosis code','Age']].corr()
```

### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(train_x[['Age', 'BS', 'SNSB']], train_y)
```

```python
pred_rfr = rfr.predict(test_x[['Age', 'BS', 'SNSB']])
pred_rfr = pred_rfr.astype('int')
pred_rfr = pred_rfr.tolist()
lt = list(target_data.index)
```

# Target 데이터 인코딩

```python
data = pd.read_csv('../알츠하이머/data/modeling.csv')
```

```python
data['Diagnosis code'] = data['Diagnosis code'].apply(lambda x: 1 if x==5 else 0)
data
```

# 1차 모델링

```python
put = data.drop('Diagnosis code', axis=1)
target = data['Diagnosis code']

from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(put, target, test_size=0.15, random_state=42)
train_x, val_x, train_y, val_y = train_test_split(train_x, train_y, test_size=0.2, random_state=42)
```

##  XGB

```python
from xgboost.sklearn import XGBClassifier
xgbc = XGBClassifier(n_estimators=150, random_state= 0)
xgbc.fit(train_x, train_y)
```

```python
val_pred = xgbc.predict(val_x)

from sklearn.metrics import accuracy_score
print(accuracy_score(val_pred, val_y))
```

```python
test_pred = xgbc.predict(test_x)

from sklearn.metrics import accuracy_score
print(accuracy_score(test_pred, test_y))
```

```python
train_x, test_x, train_y, test_y = train_test_split(put, target, test_size=0.2, random_state=42)
xgbc = XGBClassifier()
xgbc.fit(train_x, train_y)
```

```python
pred_xgbc = xgbc.predict(test_x)

from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score, confusion_matrix
print('accuracy_score:',accuracy_score(test_y,pred_xgbc))
print('f1_score',f1_score(test_y,pred_xgbc,average='macro'))
print('recall_score',recall_score(test_y,pred_xgbc,average='macro'))
print('precision_score', precision_score(test_y,pred_xgbc,average='macro'))
print('confusion_matrix',confusion_matrix(test_y,pred_xgbc))
```
