```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

i = ['GDS', 'CDR-SOB', 'positivity', '(SUV)PUT-l',
       '(SUV)PUT-r', '(SUV)GCA-r', '(SUV)GCA-l', '(SUV)PQ-r', '(SUV)GCP-r',
       '(SUV)GCP-l', '(SUV)LTC-r', '(SUV)PQ-l', '(SUV)FC-r', '(SUV)LTC-l',
       '(SUV)CN-r', '(SUV)PC-r', '(SUV)FC-l', 'CDR', '(SUV)CN-l', '(SUV)PC-l',
       '(SUV)OC-l', '(SUV)OC-r', 'APOE', '(Vol)LTC-l', '(Vol)PQ-r',
       '(Vol)PQ-l', '(Vol)PC-l', '(Vol)MTC-l', '(Vol)MTC-r', 'MMSE']
inpu = df3[i]
target = df3['Diagnosis code']

train_x, val_x, train_y, val_y = train_test_split(inpu, target, test_size=0.2, shuffle=True, stratify=target)
```

```python
rfc = RandomForestClassifier(n_estimators=150, max_depth=7)
rfc.fit(train_x, train_y)
```

```python
pred = rfc.predict(val_x)
from sklearn.metrics import accuracy_score, roc_auc_score
print(accuracy_score(val_y, pred))
print(roc_auc_score(val_y, pred))
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(rfc, inpu, target, cv=10)
print(scores.mean())
```

## 그리드서치

```python
from sklearn.model_selection import GridSearchCV

# 데이터 준비
inpu = df3[i]
target = df3['Diagnosis code']

# 모델 선택
model = RandomForestClassifier(crietrion='entropy')

# 그리드 정의
param_grid = {
    'n_estimators': [75,100,125,150,175],  # 첫 번째 하이퍼파라미터 값들의 리스트
    'max_depth': [None,7, 8, 9, 10,20],  # 두 번째 하이퍼파라미터 값들의 리스트
}

# 그리드 서치 수행
grid_search = GridSearchCV(model, param_grid, cv=10)  # 5-겹 교차 검증을 사용
grid_search.fit(inpu, target)

# 최적의 하이퍼파라미터 조합과 성능 출력
print("최적의 하이퍼파라미터:", grid_search.best_params_)
print("최고 성능:", grid_search.best_score_)
```
